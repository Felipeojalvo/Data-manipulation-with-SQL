```python
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
```

```python
data = """Series_or_movies_Title,No_of_Votes (Count)
"The Shawshank Redemption",2343110
"The Dark Knight",2303232
Inception,2067042
"Fight Club",1854740
"Pulp Fiction",1826188
"Forrest Gump",1809221
"The Matrix",1676426
"The Lord of the Rings: The Fellowship of the Ring",1661481
"The Lord of the Rings: The Return of the King",1642758
"The Godfather",1620367
"""
```

```python
df = pd.read_csv(StringIO(data))
```

```python
df['Series_or_movies_Title'] = df['Series_or_movies_Title'].apply(lambda x: x.split(', '))
```

```python
df_exploded = df.explode('No_of_Votes (Count)')
```

```python
df_exploded['Series_or_movies_Title'] = df_exploded['Series_or_movies_Title'].apply(lambda x: x[0] if isinstance(x, list) else x)
pivot_table = df_exploded.groupby('Series_or_movies_Title')['No_of_Votes (Count)'].sum().reset_index()
```

```python
pivot_table_sorted = pivot_table.sort_values(by='No_of_Votes (Count)', ascending=False)
```

```python
print(pivot_table_sorted.to_string(index=False))
```

```python
# Create DataFrame from text string
df = pd.read_csv(StringIO(data))

```

```python
# Convert rows with multiple genders to lists
df['Genre'] = df['Genre'].apply(lambda x: x.split(', '))

```

```python
# Explode to separate lists into individual rows
df_exploded = df.explode('Genre')

```

```python
# Calculate the sum of films by genre
pivot_table = df_exploded.groupby('Genre')['Number_of_Movies (Count)'].sum().reset_index()

```

```python
# Sort the table by the number of movies in descending order
pivot_table_sorted = pivot_table.sort_values(by='Number_of_Movies (Count)', ascending=False)


```

```python
# Print the sorted table
print(pivot_table_sorted.to_string(index=False))
```

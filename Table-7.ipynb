```python
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
```

```python
data = """Certificate,Average_Duration
G,92.50
Passed,103.00
TV-MA,106.00
PG,112.67
16,130.00
R,130.83
TV-14,132.00
PG-13,133.50
U,134.62
UA,138.33
A,142.03
"""
```

```python
df = pd.read_csv(StringIO(data))
```

```python
df['Certificate'] = df['Certificate'].apply(lambda x: x.split(','))
```

```python
df_exploded = df.explode('Average_Duration')
```

```python
df_exploded['Certificate'] = df_exploded['Certificate'].apply(lambda x: x[0] if isinstance(x, list) else x)
pivot_table = df_exploded.groupby('Certificate')['Average_Duration'].sum().reset_index()
```

```python
pivot_table_sorted = pivot_table.sort_values(by='Average_Duration', ascending=False)
```

```python
print(pivot_table_sorted.to_string(index=False))
```

```python
# Create DataFrame from text string
df = pd.read_csv(StringIO(data))
```

```python
df['Released_Year'] = df['Released_Year'].astype(str)
df['Released_Year'] = df['Released_Year'].apply(lambda x: x.split(','))
```

```python
df_exploded = df.explode('Movies (Count)')
```

```python
df_exploded['Released_Year'] = df_exploded['Released_Year'].apply(lambda x: x[0] if isinstance(x, list) else x)
pivot_table = df_exploded.groupby('Released_Year')['Movies (Count)'].sum().reset_index()
```

```python
pivot_table_sorted = pivot_table.sort_values(by='Movies (Count)', ascending=False)
```

```python
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
```

```python
data = """Series_Title,Gross
"Sherlock Jr.","977,375"
"Schindler's List","96,898,818"
RashÃ´mon,"96,568"
"Bacheha-Ye aseman","933,933"
"OMG: Oh My God!","923,221"
Haider,"901,610"
Chhichhore,"898,575"
"Avengers: Endgame","858,373,000"
"Song of the Sea","857,524"
Aliens,"85,160,248"
"""
```

```python
df = pd.read_csv(StringIO(data))
```

```python
df['Series_Title'] = df['Series_Title'].apply(lambda x: x.split(','))
```

```python
df_exploded = df.explode('Gross')
```

```python
df_exploded['Series_Title'] = df_exploded['Series_Title'].apply(lambda x: x[0] if isinstance(x, list) else x)
pivot_table = df_exploded.groupby('Series_Title')['Gross'].sum().reset_index()
```

```python
pivot_table_sorted = pivot_table.sort_values(by='Gross', ascending=False)
```

```python
print(pivot_table_sorted.to_string(index=False))
```
